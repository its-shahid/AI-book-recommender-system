#AI BOOK RECOMMENDER SYSYTEM

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python

books = pd.read_csv('books.csv')
ratings = pd.read_csv('ratings.csv')
users = pd.read_csv('users.csv')
```

```python
books = books.drop(columns=['Image-URL-S', 'Image-URL-M'])
```

```python

books.head()
```

```python
books.rename(columns={"Book-Title":'title',
                      'Book-Author':'author',
                     "Year-Of-Publication":'year',
                     "Publisher":"publisher",
                     "Image-URL-L":"image_url"},inplace=True)
```

```python

ratings.head()
```

```python
ratings.shape
```

```python
ratings.rename(columns={"User-ID":'user_id',
                      'Book-Rating':'rating'},inplace=True)
```

```python
x = ratings['user_id'].value_counts() > 200
```

```python

x[x].shape
```

```python

y = x[x].index
```

```python

y
     
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python

ratings.head()
```

```python

ratings_with_books = ratings.merge(books, on='ISBN')
```

```python

ratings_with_books.head()
```

```python
number_rating = ratings_with_books.groupby('Book-Title')['rating'].count().reset_index()
```

```python

number_rating
```

```python

number_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
```

```python

final_rating = ratings_with_books.merge(number_rating, on='Book-Title')
```

```python

final_rating.head()
```

```python
final_rating = final_rating[final_rating['num_of_rating'] >= 50]
```

```python

final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','Book-Title'], inplace=True)
     
```

```python

final_rating.shape
```

```python


book_pivot = final_rating.pivot_table(
    columns='user_id',
    index='Book-Title',
    values='rating'
)
```

```python

book_pivot.head(5)
```

```python

book_pivot.shape
```

```python

book_pivot.fillna(0, inplace=True)
```

```python
book_pivot.head()
```

Training Model

```python

from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute', metric='cosine')
```

```python

model.fit(book_pivot)
```

```python

_ , suggestion = model.kneighbors(book_pivot.iloc[100,:].values.reshape(1,-1), n_neighbors=6)
```

```python

suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

#Finding a poster link

```python

book_name = []
for book_id in suggestion:
    book_name.append(book_pivot.index[book_id])
```

```python

book_name[0]
```

```python


ids_index = [final_rating[final_rating['Book-Title'] == name].index[0] for name in book_name[0] if not final_rating[final_rating['Book-Title'] == name].empty]
```

```python
for idx in ids_index:
    url = final_rating.iloc[idx]['Image-URL-L']
    print(url)
```

```python

 book_names= list(book_pivot.index)
     
```

```python

import pickle
pickle.dump(model,open('model.pkl','wb'))
pickle.dump(book_names,open('book_names.pkl','wb'))
pickle.dump(final_rating,open('final_rating.pkl','wb'))
pickle.dump(book_pivot,open('book_pivot.pkl','wb'))
```

Testing model

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    _ , suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6)
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python

book_name = "1984"
recommend_book(book_name)
```
